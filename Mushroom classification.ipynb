# Applying Decision Tree algorithm:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
Decision=DecisionTreeClassifier(criterion='entropy')
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.0973,shuffle=True)
     

Decision.fit(X_train,Y_train)
     
DecisionTreeClassifier(criterion='entropy')

forecast=Decision.predict(X_test)
     
```

```python
print(accuracy_score(Y_test,forecast))
```

```python
h=pd.DataFrame(np.c_[Y_test,forecast],columns=["Acutual","Prediction"])
     
```
